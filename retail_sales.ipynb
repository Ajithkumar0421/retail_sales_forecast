```python
y = df1["Weekly_Sales"]
x = df1.drop(["Weekly_Sales","Date","Size",'Week','Markdown'],axis = 1)
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import cross_val_score, KFold
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20)
```

```python
model = RandomForestRegressor(n_estimators=40)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```python
nfolds=5
kf = KFold(n_splits=nfolds, shuffle=True, random_state=42)
```

```python
#perform cross validation:
cv_scores_r2 = cross_val_score(model, x,y, cv=kf, scoring="r2")
print("cross validation R-squared score is:", cv_scores_r2)

#calculate the average R_squared score:
avg_cv_score_r2 = cv_scores_r2.mean()
print("Average cross validation R-squared score is:", avg_cv_score_r2)
```

```python
#initialize the base estimator:
base_estimator = DecisionTreeRegressor(max_depth=100)
#initialize adaboostregressor:
model1 = AdaBoostRegressor(base_estimator=base_estimator)
```

```python
num_folds = 5
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)
#perform cross_validation:
cv_scores_r2 = cross_val_score(model1, x,y, cv=kf, scoring="r2")
print("Cross-Validation R-squared Scores:", cv_scores_r2)

# Calculate the average R-squared score
avg_cv_score_r2 = cv_scores_r2.mean()
print("Average Cross-Validation R-squared Score:", avg_cv_score_r2)
  
```

```python
model2 = GradientBoostingRegressor()
num_f = 5
kf = KFold(n_splits=num_f, shuffle= True, random_state=42)
#cross_validation:
cv_scores_r2 = cross_val_score(model2, x,y, cv=kf, scoring="r2")
# Display the cross-validation R-squared scores
print("Cross-Validation R-squared Scores:", cv_scores_r2)

# Calculate the average R-squared score
avg_cv_score_r2 = cv_scores_r2.mean()
print("Average Cross-Validation R-squared Score:", avg_cv_score_r2)
```
